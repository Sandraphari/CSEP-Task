```python
import os
import time
import random
import selenium
import logging
import boto3
import pandas as pd
from datetime import datetime,date
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import WebDriverException
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(executable_path=ChromeDriverManager().install())
```

```python
column_name=['Date','Hour','Time Block','A1','A2','E1','E2','N1','N2','N3','S1','S2','S3','W1','W2','W3','MCP']
df1_dec=pd.DataFrame(columns=column_name)
df1_dec
```

```python
for i in range(31):
    df1_dec_new=pd.DataFrame(columns=column_name)
    page_content = driver.page_source
    page_soup = BeautifulSoup(page_content, 'html.parser')
    div = page_soup.find('div',{'id':'VisibleReportContentctl00_InnerContent_reportViewer_ctl09'})

    for j in range(96):
        df1_dec_new.at[j,'Date']=div.find_all('tr')[9].find_all('td')[36].text
    c=0
    while c <96:
        for k in range(4):
            df1_dec_new.at[c+k,'Hour']=div.find_all('tr')[9].find_all('td')[37+(c//4)*65].text
        c+=4  
    d=0
    m=38
    while d <96:
        for k in range(2,17):
            df1_dec_new.iloc[d,k]=div.find_all('tr')[9].find_all('td')[m].text
            m=m+1
        if (d+1)%4==0:
            m+=2
        else:    
            m+=1   
        d+=1 
    driver.find_element(By.XPATH,'//*[@id="ctl00_InnerContent_reportViewer_ctl05_ctl00_Next_ctl00_ctl00"]').click()
    time.sleep(20)
    df1_dec=df1_dec.append(df1_dec_new)
```

```python
df1_dec
```

```python
column_name2=['Date','','A1','A2','E1','E2','N1','N2','N3','S1','S2','S3','W1','W2','W3','MCP']
df2_dec=pd.DataFrame(columns=column_name2)
df2_dec
```

```python
df2_dec_new=pd.DataFrame(columns=column_name2)
page_content = driver.page_source
page_soup = BeautifulSoup(page_content, 'html.parser')
div = page_soup.find('div',{'id':'VisibleReportContentctl00_InnerContent_reportViewer_ctl09'})
c=0
while c<85:
    for k in range(5):
        df2_dec_new.at[c+k,'Date']=div.find_all('tr')[10].find_all('td')[35+(c//5)*81].text
    c+=5
c=85
for k in range(3):
    df2_dec_new.at[c+k,'Date']=div.find_all('tr')[10].find_all('td')[35+(c//5)*81].text  

d=0
m=36
while d <88:
    for k in range(1,16):
        df2_dec_new.iloc[d,k]=div.find_all('tr')[10].find_all('td')[m].text
        m=m+1
    if (d+1)%5==0:
        m+=2
    else:    
        m+=1   
    d+=1 
df2_dec=df2_dec.append(df2_dec_new)  
    
driver.find_element(By.XPATH,'//*[@id="ctl00_InnerContent_reportViewer_ctl05_ctl00_Next_ctl00_ctl00"]').click()
time.sleep(20)
```

```python
df2_dec_new=pd.DataFrame(columns=column_name2)
page_content = driver.page_source
page_soup = BeautifulSoup(page_content, 'html.parser')
div = page_soup.find('div',{'id':'VisibleReportContentctl00_InnerContent_reportViewer_ctl09'})
c=0
for k in range(2):
    df2_dec_new.at[c+k,'Date']=div.find_all('tr')[9].find_all('td')[35].text
c=2
while c<67:
    for k in range(5):
        df2_dec_new.at[c+k,'Date']=div.find_all('tr')[9].find_all('td')[68+((c-2)//5)*81].text  
    c+=5    
d=0
m=36
while d <67:
    for k in range(1,16):
        df2_dec_new.iloc[d,k]=div.find_all('tr')[9].find_all('td')[m].text
        m=m+1
    if (d-1)%5==0:
        m+=2
    else:    
        m+=1   
    d+=1 
df2_dec=df2_dec.append(df2_dec_new)  
    
```
